###Consulta especifica na tabela

```python
nome = Tabela_cadastro['nome']
display(nome)
```

### conhecendo metodo .loc    
- busca 1 linha de acordo com alguma condição especifica
- linhas e colunas especificas
- valor especifico

```python
display(Tabela_cadastro.loc[0 : 4])
```

```python
display(Tabela_cadastro.loc[Tabela_cadastro['nome' ] == 'Bianca'])
```

```python
dadosespecificos = Tabela_cadastro.loc[Tabela_cadastro['nome'] == 'Bianca',['nome','idade','telefone']]
display(dadosespecificos)
```

```python
print(Tabela_cadastro.loc[1, "telefone"])
```

Adicionar nova coluna

```python
Tabela_cadastro['endereco'] = 'Rua ' + Tabela_cadastro['endereco']
display(Tabela_cadastro)
```

```python
Tabela_cadastro.loc[: , 'numero'] = Tabela_cadastro['idade']
display(Tabela_cadastro)
```
