Importações

```python
import pandas as pd
```

```python
vendas = pd.read_excel('Vendas.xlsx')
display(vendas)
```

```python
#buscando valores repetidos na coluna id loja   
```

```python
iguatemi = vendas.loc[vendas['ID Loja'] == 'Iguatemi Esplanada']
morumbi = vendas.loc[vendas['ID Loja'] == 'Shopping Morumbi']
norte = vendas.loc[vendas['ID Loja'] == 'Norte Shopping']
ribeirao = vendas.loc[vendas['ID Loja'] == 'Ribeirão Shopping']
fortaleza = vendas.loc[vendas['ID Loja'] == 'Rio Mar Shopping Fortaleza']
bourbon = vendas.loc[vendas['ID Loja'] == 'Bourbon Shopping SP']
uberlandia = vendas.loc[vendas['ID Loja'] == 'Center Shopping Uberlândia']
campinas = vendas.loc[vendas['ID Loja'] == 'Iguatemi Campinas']


```

```python
with pd.ExcelWriter("Iguatemi Esplanada.xlsx", mode="w", engine="openpyxl") as writer:
    iguatemi.to_excel(writer) 
with pd.ExcelWriter("Shopping Morumbi.xlsx", mode="w", engine="openpyxl") as writer:
    morumbi.to_excel(writer) 
with pd.ExcelWriter("Norte Shopping.xlsx", mode="w", engine="openpyxl") as writer:
    norte.to_excel(writer)    
with pd.ExcelWriter("Ribeirão Shopping.xlsx", mode="w", engine="openpyxl") as writer:
    ribeirao.to_excel(writer)
with pd.ExcelWriter("Rio Mar Shopping Fortaleza.xlsx", mode="w", engine="openpyxl") as writer:
    fortaleza.to_excel(writer)
with pd.ExcelWriter("Bourbon Shopping SP.xlsx", mode="w", engine="openpyxl") as writer:
    bourbon.to_excel(writer)
with pd.ExcelWriter("Center Shopping Uberlândia.xlsx", mode="w", engine="openpyxl") as writer:
    uberlandia.to_excel(writer)
with pd.ExcelWriter("Iguatemi Campinas.xlsx", mode="w", engine="openpyxl") as writer:
    campinas.to_excel(writer)    
```
